```python
#### task 4 ######
      #### 4a #####
cursor.execute('''SELECT Time.timestamp, Weather.temp, CarSharing.demand
                        FROM Time
                        INNER JOIN Weather ON Time.id = Weather.id
                        INNER JOIN CarSharing ON Time.id = CarSharing.id
                        WHERE Weather.temp = (SELECT MIN(temp) From Weather);
''')

print([dec[0] for dec in cursor.description])
result=cursor.fetchall()
for i in result: 
        print(i)
```

```python
    ##### 4b ####
   
cursor.execute('''SELECT 
                      Holiday.workingday,
                      AVG(windspeed) AS avg_windspeed,
                      MAX(windspeed) AS highest_windspeed,
                      MIN(windspeed) AS lowest_windspeed,
                      AVG(humidity) AS avg_humidity,
                      MAX(humidity) AS highest_humidity,
                      MIN(humidity) AS lowest_humidity
                      FROM
                      Weather
                      INNER JOIN Time ON Weather.id=Time.id
                      INNER JOIN Holiday ON Time.id=Holiday.id
                      WHERE
                      strftime('%Y', Time.timestamp)='2017'
                      GROUP BY 
                      Holiday.workingday;
''')
print([dec[0] for dec in cursor.description])
result=cursor.fetchall()
for i in result: 
        print(i)
```

```python
#### 4c ###
cursor.execute('''SELECT 
                  Time.weekday_name AS weekday,
                  Time.month AS month,
                  season_name AS season,
                  AVG(CarSharing.demand) AS average_demand
                  FROM
                  CarSharing
                  INNER JOIN Time ON CarSharing.id=Time.id
                  WHERE
                  strftime('%Y', Time.timestamp)='2017'
                  GROUP BY weekday, month,season
                  ORDER BY average_demand DESC
                  LIMIT 1;
''')
print([dec[0] for dec in cursor.description])
result=cursor.fetchall()
for i in result: 
        print(i)
```

```python
  #### 4d ####
cursor.execute('''SELECT 
                  Weather.humidity_category,
                  AVG(CarSharing.demand) AS average_demand
                  FROM
                  CarSharing
                  INNER JOIN Time ON CarSharing.id=Time.id
                  INNER JOIN Weather ON CarSharing.id=Weather.id
                  WHERE strftime('%Y', Time.timestamp)='2017'
                  AND Weather.humidity_category IS NOT NULL
                  GROUP BY Weather.humidity_category
                  ORDER BY average_demand DESC;
''')
print([dec[0] for dec in cursor.description])
result=cursor.fetchall()
for i in result: 
        print(i)
```
